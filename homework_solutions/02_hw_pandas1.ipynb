```python
import pandas as pd

df = pd.read_csv("../data/flights.csv", engine="pyarrow")
```

```python
# 1
df['WEATHER_DELAY'].max()
```

```python
# 2
# df.query("ORIGIN_AIRPORT == 'LAX' and MONTH == 12 and DESTINATION_AIRPORT=='ATL'")['ARRIVAL_DELAY'].max()

# 2
q2 = (
    df['ORIGIN_AIRPORT'] == 'LAX') & \
    (df['MONTH'] == 12) & \
    (df['DESTINATION_AIRPORT'] == 'ATL'
)
df.loc[q2, 'ARRIVAL_DELAY'].max()
```

```python
# 3
# df3 = df.query("ORIGIN_AIRPORT == 'LAX' and MONTH == 12 and DESTINATION_AIRPORT=='ATL'")
# df3.iloc[df3['ARRIVAL_DELAY'].argmax(), :]['TAIL_NUMBER']

# 3
df3 = df.loc[q2, :]
df3.loc[df3['ARRIVAL_DELAY'] == df3['ARRIVAL_DELAY'].max(), 'TAIL_NUMBER']
```

```python
# 4
df3['ARRIVAL_DELAY'].min()
```

```python
# 5
df3.loc[df3['ARRIVAL_DELAY'] == df3['ARRIVAL_DELAY'].min(), :]['TAIL_NUMBER']
```

```python
# 6,7,8
# df4 = df.query("ORIGIN_AIRPORT == 'HNL'")
df4 = df.loc[df['ORIGIN_AIRPORT'] == 'HNL']
(
    df4
    .iloc[df4['AIR_TIME'].argmax(), :]
    .loc[["AIR_TIME", "DESTINATION_AIRPORT", "AIRLINE"]]
)
```

```python
# 9, 10, 11, 12
apts = ('ORD', 'AUS')
# df5 = df.query("ORIGIN_AIRPORT in @apts and DESTINATION_AIRPORT in @apts")

# q9 = df['ORIGIN_AIRPORT'].isin(apts) & df['DESTINATION_AIRPORT'].isin(apts) 
q9 = (
    (df['ORIGIN_AIRPORT'] == "ORD") & (df['DESTINATION_AIRPORT'] == 'AUS') |
    (df['ORIGIN_AIRPORT'] == "AUS") & (df['DESTINATION_AIRPORT'] == 'ORD')
)
df5 = df.loc[q9, :]
df5['AIR_TIME'].min(), df5['AIR_TIME'].mean(), df5['AIR_TIME'].median(), df5['AIR_TIME'].max()
```

```python
# 13, 14
# df.query("ORIGIN_AIRPORT == 'YUM'")['AIRLINE'].value_counts()
df.loc[df['ORIGIN_AIRPORT'] == 'YUM', 'AIRLINE'].value_counts()
```

```python
# 13, 14
# This is even better
# df.query("ORIGIN_AIRPORT == 'YUM'")['AIRLINE'].value_counts()
df.loc[
    (df['ORIGIN_AIRPORT'] == 'YUM') & (df['CANCELLED'] == 0),
    'AIRLINE'].value_counts()
```

```python
# 15, 16
df.dropna(subset="AIR_TIME").sort_values("AIR_TIME").head(20)
# PSG	WRG	
```
